```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('mineral water t1.txt', delim_whitespace=True, header=None)

data = np.array(data, dtype=float)

x = data[:,0]
y = data[:,1]

plt.figure(figsize=(10,6))

plt.plot(x, y)

plt.xlabel('Decay Time(ms)')
plt.ylabel('Voltage (V)')
plt.title('Mineral Water Spin-Lattice Relaxation')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# Define the exponential decay function
def exponential_decay(t, A, T1):
    return A * np.exp(-t / T1)

# Load the data from the file
data = np.loadtxt('mineral water t1.txt')
time = data[:, 0]
voltage = data[:, 1]

# Fit the exponential decay function to the data
initial_guess = [max(voltage), 10]  # Initial guess for A and T1
params, covariance = curve_fit(exponential_decay, time, voltage, p0=initial_guess)
A_fit, T1_fit = params

# Generate data points for the fitted curve
time_fit = np.linspace(min(time), max(time), 100)
voltage_fit = exponential_decay(time_fit, A_fit, T1_fit)

# Plot the data and the fitted curve
plt.figure(figsize=(10, 6))
plt.scatter(time, voltage, label='Recorded Data', color='blue')
plt.plot(time_fit, voltage_fit, label=f'Exponential Fit: A={A_fit:.2f}, T1={T1_fit:.2f}', color='red')
plt.xlabel('Decay Time (ms)')
plt.ylabel('Voltage (V)')
plt.title('Exponential Decay of Voltage over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

data = pd.read_csv('mineral water t1_2.txt', delim_whitespace=True, header=None)

data = np.array(data, dtype=float)

x = data[:,0]
y = data[:,1]

def M_model(tau, M_0, T_1):
    return M_0 * (1 - 2 * np.exp(-tau / T_1))

# Example synthetic data
tau_data = np.array(x)
M_data = np.array(y)

from scipy.optimize import curve_fit

# Fit the model to the data
popt, pcov = curve_fit(M_model, tau_data, M_data)

# Extract the fitted parameters
M_0_fit, T1_fit = popt


# Print the fitted parameters
print(f"Fitted M_0: {M_0_fit}")
print(f"Fitted T1: {T1_fit}")

import matplotlib.pyplot as plt

# Generate points for the fitted curve
tau_fit = np.linspace(0, max(tau_data), 100)
M_fit = M_model(tau_fit, *popt)

# Plotting
plt.scatter(tau_data, M_data, label='Experimental Data')
plt.plot(tau_fit, M_fit, color='red')
plt.xlabel('Tau')
plt.ylabel('M(z, Tau)')
plt.legend()
plt.title('Fitting M(z, Tau) to Model')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Load the data
data = np.loadtxt('mineral water t2.txt')
time = data[:, 0]  # Delay time (t)
voltage = data[:, 1]  # Voltage (M_xy)

# Define the model function
def model(t, M0, T2):
    return M0 * np.exp(-t / T2)

# Fit the data to the model
popt, pcov = curve_fit(model, time, voltage, p0=(3, 0.01))

# Extract the fitting parameters
M0, T2 = popt
print(f'Fitted M0: {M0}')
print(f'Fitted T2: {T2}')

# Generate fitted data for plotting
time_fit = np.linspace(min(time), max(time), 100)
voltage_fit = model(time_fit, M0, T2)

# Plot the data and the fitted curve
plt.figure(figsize=(8, 6))
plt.scatter(time, voltage, label='Data', color='blue')
plt.plot(time_fit, voltage_fit, label='Fitted Curve', color='red')
plt.xlabel('Delay Time (ms)')
plt.ylabel('Voltage (V)')
plt.title('Mineral Water Spin-Spin Relaxation')
plt.legend()
plt.show()
```
