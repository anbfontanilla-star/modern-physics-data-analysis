```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

data = pd.read_csv('SampleData_Summer2024.dat', delim_whitespace=True, header=None)

t = data.iloc[:, 0]

x1 = data.iloc[:, 1]
y1 = data.iloc[:, 2]

x2 = data.iloc[:, 4]
y2 = data.iloc[:, 5]

x3 = data.iloc[:, 7]
y3 = data.iloc[:, 8]

x4 = data.iloc[:, 10]
y4 = data.iloc[:, 11]

x5 = data.iloc[:, 13]
y5 = data.iloc[:, 14]

x6 = data.iloc[:, 16]
y6 = data.iloc[:, 17]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x1, y1, t, marker='o', linestyle='-', color='r', label='Particle 1')
ax.plot(x2, y2, t, marker='o', linestyle='-', color='b', label='Particle 2')
ax.plot(x3, y3, t, marker='o', linestyle='-', color='g', label='Particle 3')
ax.plot(x4, y4, t, marker='o', linestyle='-', color='c', label='Particle 4')
ax.plot(x5, y5, t, marker='o', linestyle='-', color='m', label='Particle 5')
ax.plot(x6, y6, t, marker='o', linestyle='-', color='y', label='Particle 6')

ax.set_title('3D Spaghetti Plot of Particles 1-6')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Time')
ax.legend()

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm

data = pd.read_csv('SampleData_Summer2024.dat', delim_whitespace=True, header=None)

x1 = data.iloc[:, 1]
x2 = data.iloc[:, 4]

fig, ax = plt.subplots()

def gaussian(x, mu, sigma):
    return norm.pdf(x, mu, sigma)

hist1, bin_edges1 = np.histogram(x1, bins=50, density=True)
bin_centers1 = (bin_edges1[:-1] + bin_edges1[1:]) / 2
popt1, _ = curve_fit(gaussian, bin_centers1, hist1)

hist2, bin_edges2 = np.histogram(x2, bins=50, density=True)
bin_centers2 = (bin_edges2[:-1] + bin_edges2[1:]) / 2
popt2, _ = curve_fit(gaussian, bin_centers2, hist2)

ax.hist(x1, bins=50, density=True, alpha=0.6, color='blue', label='Particle 1')
ax.hist(x2, bins=50, density=True, alpha=0.6, color='orange', label='Particle 2')

x = np.linspace(min(np.concatenate((x1, x2))), max(np.concatenate((x1, x2))), 1000)
ax.plot(x, gaussian(x, *popt1), 'blue', label='Gaussian Fit Particle 1')
ax.plot(x, gaussian(x, *popt2), 'orange', label='Gaussian Fit Particle 2')

ax.set_xlabel('X Position')
ax.set_ylabel('Frequency')
ax.set_title('Histogram of X Position for Particle 1 and Particle 2')
ax.legend()

plt.show()
```
