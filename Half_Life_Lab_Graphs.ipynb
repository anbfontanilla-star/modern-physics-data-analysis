```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data1 = pd.read_csv('steadybarium_2010', delim_whitespace=True, header=None)
data2 = pd.read_csv('steadbarium_2016', delim_whitespace=True, header=None)
data3 = pd.read_csv('steadybarium_2021', delim_whitespace=True, header=None)

data1 = np.array(data1, dtype=float)
data2 = np.array(data2, dtype=float)
data3 = np.array(data3, dtype=float)

x1 = data1[:,0]
x2 = data2[:,0]
x3 = data3[:,0]

y1 = data1[:,1]
y2 = data2[:,1]
y3 = data3[:,1]

plt.figure(figsize=(10,6))

plt.plot(x1, y1, label='2010')
plt.plot(x2, y2, label='2016')
plt.plot(x3, y3, label='2021')

plt.xlabel('Time(s)')
plt.ylabel('Activity of Cs-137')
plt.title('Cs-137 from Different Years')
plt.grid(True)
plt.legend()

plt.show()
```
